```python
from src.mesh_gen.vec3 import Vec3, Interpolada
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

xs = [ Vec3(xi, yi, zi) for xi, yi, zi in zip(x,y,z) ]
curvaInterpolada = Interpolada( xs )

for t in np.linspace(0,1,100):
    ax.plot( *curvaInterpolada[t].toNumpy(), marker='.')

plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
curv = curvaInterpolada.curvatura()

color = cm.rainbow( curv.evaluarLista( np.linspace(0,1,100)) )

for t,c in zip( np.linspace(0,1,100), color):
    ax.plot( *curvaInterpolada[t].toNumpy(), color=c, marker='.')
plt.show()
```
